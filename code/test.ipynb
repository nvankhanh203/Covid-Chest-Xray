```python
import tensorflow as tf
import cv2
import numpy as np

# load img
def load_img_to_predict(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_COLOR) # read color img
    img = cv2.resize(img, (224, 224)) # resize img to same model size
    img = img/255.0 # img normalized
    return img.reshape(1, 224, 224, 3)

def load_model(model_path):
    return tf.keras.models.load_model(model_path)


MODEL_PATH = r"F:\Covid Chest-Xray\model\vgg16_modify_last_state.keras" # link model
IMG_PATH = r"F:\Covid Chest-Xray\split_data\test\normal\Normal-7.png" # link img need to predict

LABELS = {
    0: "covid",
    1: "normal"
}

model = load_model(MODEL_PATH)
img = load_img_to_predict(IMG_PATH)

output = model.predict(img, verbose=1)[0, 0]
index = int(np.round(output))
print(f"Probability: {output}")
print(f"Classify: {LABELS[index]}")

```
